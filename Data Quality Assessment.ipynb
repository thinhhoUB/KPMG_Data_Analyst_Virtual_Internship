```python
mapping= {"NSW":"NSW","New South Wales":"NSW","Victoria":"VIC","VIC":"VIC","QLD":"QLD"}
```

```python
cust_address['state']= cust_address['state'].replace(mapping)
cust_address['state'].unique()
```
